Features Selection Column

```python
exercises.isnull().count()
```

```python
exercises=exercises[['ID','Title','Description','Type','BodyPart','Equipment','Level','URL','Area','Description']]
```

```python
exercises
```

```python
exercises['tags']=exercises['Type']+exercises['BodyPart']+exercises['Equipment']
```

```python
exercises
```

```python
new_data=exercises.drop(columns=['Type','BodyPart','Equipment'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=42, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['Level'].values.astype('U')).toarray()
```

```python
vector.shape
```

Recommendation System with Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['Level']=="Beginner"].index[0]
```

```python
new_data[new_data['Level']=="Intermediate"].index[0]
```

```python
new_data[new_data['Level']=="Expert"].index[0]
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].Level)
```

```python
def recommend(exelevel):
    index=new_data[new_data['Level']==exelevel].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].Title)
```

```python
recommend("Beginner")
```
